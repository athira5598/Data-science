```python
combine=train.append(test,ignore_index=True)
combine.shape
```

```python
def remove_pattern(input_txt,pattern):
    r=re.findall(pattern,input_txt)
    for word in r:
        input_txt=re.sub(word,'',input_txt)
    return input_txt
```

```python
combine['new_tweet']=np.vectorize(remove_pattern)(combine['tweet'],"@[\w]*")
combine.head(10)
```

```python
combine["new_tweet"]=combine["new_tweet"].apply(lambda x: ''.join([w for w in x.split() if len(w)>3]))
combine.head()
```

```python
combine['new_tweet']=combine['new_tweet'].str.replace("[^a-zA-Z#]"," ")
combine.head(10)
```

```python
combine['new_tweet']=combine['new_tweet'].apply(lambda x:''.join([w for w in x.split() if len(w)>3]))
combine.head(10)
```

```python
token_tweet=combine['new_tweet'].apply(lambda x:x.split())
token_tweet.head(10)
```

```python
from nltk.stem.porter import *
stemmer=PorterStemmer()
token_tweet=token_tweet.apply(lambda x: [stemmer.stem(i)for i in x])
```

```python
for i in range(len(token_tweet)):
    token_tweet[i]=' '.join(token_tweet[i])
combine['new_tweet']=token_tweet
```

```python
all_words=' '.join([text for text in combine['new_tweet']]) 
from wordcloud import WordCloud
wordcloud=WordCloud(width=800,height=500,random_state=21,max_font_size=110).generate(all_words)
plt.figure(figsize=(10,7))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
normal_word=' '.join([text for text in combine['new_tweet'][combine['label']==0]])
wordcloud=WordCloud(width=800,height=500,random_state=21,max_font_size=110).generate(normal_word)
plt.figure(figsize=(10,7))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
negative_word=' '.join([text for text in combine['new_tweet'][combine['label']==1]])
wordcloud=WordCloud(width=800,height=500,random_state=21,max_font_size=110).generate(negative_word)
plt.figure(figsize=(10,7))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def hashtag_extract(x):
    hashtags=[]
    for i in x:
        ht=re.findall(r"#(\w+)", i)
        hashtags.append(ht)
    return hashtags
ht_regular=hashtag_extract(combine['new_tweet'][combine['label']==0])
ht_negative=hashtag_extract(combine['new_tweet'][combine['label']==1])
ht_regular=sum(ht_regular,[])
ht_negative=sum(ht_negative,[])                      
```

```python
a=nltk.FreqDist(ht_regular)
d=pd.DataFrame({'Hashtag':list(a.keys()),'Count':list(a.values())})
d=d.nlargest(columns="Count",n=20)
plt.figure(figsize=(16,5))
ax=sns.barplot(data=d,x="Hashtag",y="Count")
ax.set(ylabel='Count')
plt.show()
```

```python
b=nltk.FreqDist(ht_negative)
e=pd.DataFrame({'Hashtag':list(b.keys()),'Count':list(b.values())})
e=e.nlargest(columns="Count",n=20)
plt.figure(figsize=(16,5))
ax=sns.barplot(data=e,x="Hashtag",y="Count")
ax.set(ylabel='Count')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import gensim
bow_vectorizer=CountVectorizer(max_df=0.90,min_df=2,max_features=1000,stop_words='english')
bow=bow_vectorizer.fit_transform(combine['new_tweet'])
bow.shape
```

```python
tfidf_vectorizer=TfidfVectorizer(max_df=0.90,min_df=2,max_features=1000,stop_words='english')
tfidf=tfidf_vectorizer.fit_transform(combine['new_tweet'])
tfidf.shape
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
```

```python
train_bow=bow[:31962,:]
test_bow=bow[31962:,:]
xtrain_bow,xvalid_bow,ytrain,yvalid=train_test_split(train_bow,train['label'],random_state=42,test_size=0.3)
```

```python
train_bow.shape,test_bow.shape
```

```python
Ireg=LogisticRegression()
Ireg.fit(xtrain_bow,ytrain)
prediction=Ireg.predict_proba(xvalid_bow)
prediction_int=prediction[:,1]>=0.3
prediction_int=prediction_int.astype(np.int)
f1_score(yvalid,prediction_int)
```

```python
test_pred=Ireg.predict_proba(test_bow)
test_pred_int=test_pred[:,1]
test_pred_int=test_pred_int.astype(np.int)
test['label']=test_pred_int
submission=test[['id','label']]
submission.to_csv("sub_Ireg_bow.csv",index=False)
```

```python
train_tfidf=tfidf[:31962,:]
test_tfidf=tfidf[31962:,:]
xtrain_tfidf=train_tfidf[ytrain.index]
xvalid_tfidf=train_tfidf[yvalid.index]
Ireg.fit(xtrain_tfidf,ytrain)
prediction=Ireg.predict_proba(xvalid_tfidf)
prediction_int=prediction[:,1]>=0.3
prediction_int.astype(np.int)
f1_score(yvalid,prediction_int)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=400,random_state=11).fit(xtrain_bow,ytrain)
prediction=rf.predict(xvalid_bow)
f1_score(yvalid,prediction)
```

```python
test_pred=rf.predict(test_bow)
test['label']=test_pred
submission=test[['id','label']]
submission.to_csv('sub_rf_bow.csv',index=False)
```

```python
rf=RandomForestClassifier(n_estimators=400,random_state=11).fit(xtrain_tfidf,ytrain)
prediction=rf.predict(xvalid_tfidf)
f1_score(yvalid,prediction)
```
